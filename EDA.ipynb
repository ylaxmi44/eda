```python
#merging the first name, middle name, and last name column
salaries_archive['full_name'] = salaries_archive['first_name'].map(str) + ' ' + salaries_archive['middle_initial'].map(str)+ ' ' + salaries_archive['last_name'].map(str)


salaries_archive
```

<h3><b>Deleting coloumns</b></h3>


```python
#dropping the column first name, last name and middle name colouwn
salaries_archive=salaries_archive.drop('first_name', axis=1)

```

```python
salaries_archive= salaries_archive.drop('middle_initial', axis=1)
salaries_archive
```

```python
salaries_archive=salaries_archive.drop('last_name', axis=1)
salaries_archive
```

<b> shifting Column to first</b>

```python
# shift column 'full_name' to first position
first_column = salaries_archive.pop('full_name')
  
# insert column using insert(position,column_name,first_column) function
salaries_archive.insert(0, 'full_name', first_column)
```
